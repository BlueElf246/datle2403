##Cleaning the data

###Creating our Dataset

```python
main_data.columns
```

```python
main_data['Indicator Name'].unique()
```

```python
main_data_unem = main_data[main_data['Indicator Name'] == 'Unemployment, female (% of female labor force) (modeled ILO estimate)']
main_data_unem.head()
```

```python
main_data_unem = main_data_unem[['Country Name', 'Country Code', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']]
main_data_unem.head()
```

```python
country_data.columns
```

```python
country_data.rename(columns={"AFW": "Country Code"}, inplace=True)
country_data.rename(columns={"Unnamed: 1": "Region"}, inplace=True)
country_data.rename(columns={"Unnamed: 2": "IncomeGroup"}, inplace=True)
country_data.columns
```

```python
country_data = country_data[['Country Code', 'Region', 'IncomeGroup']]
country_data.head()
```

```python
merged_data = pd.merge(main_data_unem, country_data, on = 'Country Code')
merged_data.head()
```

##Exploratory Data Analytics

```python
merged_data.isna().sum()
```

```python
merged_data_clean = merged_data.dropna()
merged_data_clean.shape
```

```python
grouped_data_income = merged_data_clean.groupby(['IncomeGroup']).mean()
grouped_data_income
```

```python
tranposed_income_group = grouped_data_income.transpose()
tranposed_income_group
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tranposed_income_group.plot(figsize = (20, 10))
plt.title('Mean % Female Unemployment')
plt.xlabel('Year')
plt.ylabel('% Employment')
```

```python
grouped_data_region = merged_data_clean.groupby(['Region']).mean()
tranposed_region = grouped_data_region.transpose()
tranposed_region.plot(figsize = (20, 10))
plt.title('Mean % Female Unemployment by Region')
plt.xlabel('Year')
plt.ylabel('% Employment')
```

##ANOVA
Kiểm định ANOVA (ANOVA test), hay còn được gọi là phân tích phương sai (Analysis of Variance), là một kỹ thuật thống kê tham số được sử dụng để phân tích sự khác nhau giữa giá trị trung bình của các biến phụ thuộc với nhau (Ronald Fisher, 1918). 

```python
import scipy.stats as stats
```

```python
merged_data_clean.head()
```

```python
income_groups = merged_data_clean['IncomeGroup'].unique()
income_groups
```

```python
income_group_data = []
for i in range(len(income_groups)):
  income_group_data.append(merged_data_clean['1991'][merged_data_clean['IncomeGroup']== income_groups[i]])
```

```python
statistic, pvalue = stats.f_oneway(income_group_data[0], income_group_data[1], 
               income_group_data[2], income_group_data[3])
print('Statistic: %s, pvalue %s' %(statistic, pvalue))
```

```python
regions = merged_data_clean['Region'].unique()
print(regions)

regions_data = []
for i in range(len(regions)):
  regions_data.append(merged_data_clean['2019'][merged_data_clean['Region'] == regions[i]])

statistic, pvalue = stats.f_oneway(regions_data[0], regions_data[1], regions_data[2],
                                   regions_data[3], regions_data[4], regions_data[5],
                                   regions_data[6])
print('statistic: %s, pvalue %s' %(statistic, pvalue))

if pvalue < 0.05:
  print('Significant')
else:
  print('Not Significant')
```

```python
fig = plt.figure(figsize = (20, 10))
ax = fig.add_subplot(111)

ax.set_title("Boxplot of % Female Unemployment by Income Group")
ax.set

ax.boxplot(income_group_data, labels = income_groups, showmeans = True)

plt.xlabel("Country Income Group")
plt.ylabel("% Female Unemployment")
plt.show()
```

```python
fig = plt.figure(figsize = (20, 10))
ax = fig.add_subplot(111)

ax.set_title("Boxplot of % Female Unemployment by Region")
ax.set

ax.boxplot(regions_data, labels = regions, showmeans = True)

plt.xlabel("Region")
plt.ylabel("% Female Unemployment")
plt.show()
```
