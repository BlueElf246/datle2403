# Understanding language with the Transformer model


Transformer architecture is based on a concept called attention (self-attention mechanism). using attention would mean our model would be able to learn to focus on the parts of an input sequence that are more relevant to the sentiment.

# Understanding the self-attention mechanism

the goal of self-attention is to model the dependencies of each element in the output sequence to the input element

Firstly, we derive importance weights based on the similarity between the current element and all other element in the sequence. Secondly, we normalize the weights,which usually involve the use of softmax function. Thirdly, we use these weights in combination with corresponding sequence elements in order to compute the attention value
