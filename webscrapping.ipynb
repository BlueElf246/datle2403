```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page= requests.get("https://bieudogiavang.vn/gia-vang-sjc?").text # this is html text
```

```python
soup=BeautifulSoup(page,'lxml')
```

```python
table=soup.find('table',class_="min-w-full divide-y font-base text-sm divide-gray-200 text-gray-800")
```

```python
obj=table.find_all('tr')
```

```python
doc="""An Giang
Vũng Tàu
Phan Rang
Quy Nhơn
Bạc Liêu
Biên Hòa
Bắc Kạn
Bắc Giang
Bắc Ninh
Bến Tre
Bình Dương
Bình Định
Bình Phước
Bình Thuận
Cà Mau
Cao Bằng
Cần Thơ
Đà Nẵng
Đà  Nẵng
Đắk Lắk
Đắk Nông
Điện Biên
Đồng Nai
Đồng Tháp
Gia Lai
Hà Giang
Hà Nam
Hà Nội
Hà Tây
Hà Tĩnh
Hạ Long
Hải Dương
Hải Phòng
Hòa Bình
Hồ Chí Minh
Hậu Giang
Hưng Yên
Khánh Hòa
Kiên Giang
Kon Tum
Lai Châu
Lào Cai
Lạng Sơn
Lâm Đồng
Long An
Long Xuyên
Nam Định
Nha Trang
Nghệ An
Ninh Bình
Ninh Thuận
Miền Tây
Phú Thọ
Phú Yên
Quảng Bình
Quảng Nam
Quảng Ngãi
Quãng Ngãi
Quảng Ninh
Quảng Trị
Sóc Trăng
Sơn La
Tây Ninh
Thái Bình
Thái Nguyên
Thanh Hóa
Huế
Tiền Giang
Trà Vinh
Tuyên Quang
Vĩnh Long
Vĩnh Phúc
Yên Bái"""
lst=doc.split('\n')
for x in range(0,len(lst)):
    lst[x]=lst[x].lower()
```

```python
def split(text):
  text=text.lower()
  for x in lst:
    if x in text:
      remain=text.replace(x,'')
      return (str(x),str(remain))
```

```python
dct={}
location=[]
price_in=[]
price_out=[]
type_gold1=[]
for obj1 in obj:
  #location.append(obj1.find('td',class_='px-1 py-2 w-64').span.text.replace('\n',''))
  text=obj1.find('td',class_='px-1 py-2 w-64').span.text.replace('\n','')
  loc,type_gold=split(text)
  location.append(loc)
  type_gold1.append(type_gold)
  prices=obj1.find_all('td',class_='px-1 py-2 text-right')
  price_in.append(int(str(prices[0].find('span',class_='font-semibold tabular-nums').text).strip().replace('.','')))
  price_out.append(int(str(prices[1].find('span',class_='font-semibold tabular-nums').text).strip().replace('.','')))
dct['price_in']=price_in
dct['price_out']=price_out
dct['location']=location
dct['type_gold']=type_gold1
dataset=pd.DataFrame(dct)
```
