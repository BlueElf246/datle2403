```python
doc="""An Giang
Vũng Tàu
Phan Rang
Quy Nhơn
Bạc Liêu
Biên Hòa
Bình Phước
Bắc Kạn
Bắc Giang
Bắc Ninh
Bến Tre
Bình Dương
Bình Định
Bình Phước
Bình Thuận
Cà Mau
Cao Bằng
Cần Thơ
Đà Nẵng
Đà  Nẵng
Đắk Lắk
Đắk Nông
Điện Biên
Đồng Nai
Đồng Tháp
Gia Lai
Hà Giang
Hà Nam
Hà Nội
Hà Tây
Hà Tĩnh
Hạ Long
Hải Dương
Hải Phòng
Hòa Bình
Hồ Chí Minh
Huế
Hậu Giang
Hưng Yên
Khánh Hòa
Kiên Giang
Kon Tum
Lai Châu
Lào Cai
Lạng Sơn
Lâm Đồng
Long An
Long Xuyên
Phú Qúy
Nam Định
Nha Trang
Nghệ An
Ninh Bình
Ninh Thuận
Nha Trang
Miền Tây
Phú Thọ
Phú Yên
Quảng Bình
Quảng Nam
Quảng Ngãi
Quãng Ngãi
Quảng Ninh
Quảng Trị
Sóc Trăng
Sơn La
Tây Ninh
Thái Bình
Thái Nguyên
Thanh Hóa
Huế
Tiền Giang
Trà Vinh
Tuyên Quang
Vĩnh Long
HCM
Vĩnh Phúc
Yên Bái"""
lst=doc.split('\n')
for x in range(0,len(lst)):
    lst[x]=lst[x].lower()
```

```python
def location(text):
  text=text.lower()
  for x in lst:
    if str(x) in str(text):
      remain=text.replace(x,'')
      return str(x)
  return False
```

```python
# Timedelta function demonstration
from datetime import datetime, timedelta
ini_time_for_now = datetime.now()
 
# Calculating past date
# for two years
def ngay(n):
  day_total=[]
  day_for_html=[]
  for x in reversed(range(1, n)):
    day = ini_time_for_now - \
                        timedelta(days = x)
    day = str(day)[:10]
    n=list(str(day).split('-'))
    ntn = n[2] + '-' + n[1] + '-' + n[0]
    d=n[2]+n[1]+n[0]
    day_total.append(d)
    day_for_html.append(ntn)
  return day_total,day_for_html
```

```python

def xu_ly(ngay,d):
  
  html=ngay
  m=f'https://tygiadola.net/LichSuGiaVang?date={html}'
  page= requests.get(m).text # this is html text
  soup=BeautifulSoup(page,'lxml')
  data=soup.find('table',class_='table table-condensed table-hover table-bordered')
  k=0
  data_checl=str(data)
  if data_checl=='None':
    return k,False
  print(m)
  data=data.find_all('tr',class_='')
  dct={}
  loc=[]
  p_in=[]
  p_out=[]
  dae=[]
  for x in data:
    y=x.find_all('td')
    if len(y)==1:
      continue
    loca=str(y[0].text).strip().replace('\r\n','')
    l=['DOJI HCM','DOJI HN','PNJ HCM','PNJ Hà Nội','Phú Qúy SJC']
    if loca in l:
      continue
    loca=location(loca)
    if loca ==False:
      continue
    g=y[1].find('span')
    g.replaceWith('')
    h=y[2].find('span')
    h.replaceWith('')
    price_in=float(str(y[1].text).strip().replace(',','').replace('\r\n',''))
    price_out=float(str(y[2].text).strip().replace(',','').replace('\r\n',''))
    loc.append(loca)
    p_in.append(price_in)
    p_out.append(price_out)
    dae.append(d)
  dct['location']=loc
  dct['price_in']=p_in
  dct['price_out']=p_out
  dct['date']=dae
  dataset=pd.DataFrame(dct)
  return dataset,True
```

```python
def run(so_ngay):
  d=[]
  l1,l2=ngay(so_ngay)
  for x,y in zip(l2,l1):
    d1,d2=xu_ly(x,y)
    if d2==False:
      continue
    else:
      d.append(d1)
  result = pd.concat(d,ignore_index=True)
  return result
dataset=run(365)
```
